```python
# Ejercicio PHET Seleccionado - Graficando Rectas
# José Daniel Ramírez Acuña
# 1-1862-0485

import matplotlib.pyplot as plt
import numpy as np

# Definición de parámetros de la simulación
cantidad_rectas = int(input("Ingrese la cantidad de rectas que desea graficar: ")) # Cantidad de rectas a graficar
rango_x = (-10, 10)  # Rango del eje x
rango_y = (-10, 10)  # Rango del eje y

# Ingresar los valores de la pendiente y el término independiente (b)
pendientes = []
terminos_independientes = []
puntos = []
for i in range(cantidad_rectas):
    print(f"Ingrese los valores de la recta {i + 1}:")
    y2 = float(input("Ingrese el valor de y2: "))
    y1 = float(input("Ingrese el valor de y1: "))
    x2 = float(input("Ingrese el valor de x2: "))
    x1 = float(input("Ingrese el valor de x1: "))

    pendiente = (y2 - y1) / (x2 - x1)
    termino_independiente = y1 - pendiente * x1
    pendientes.append(pendiente)
    terminos_independientes.append(termino_independiente)
    puntos.append(((x1, y1), (x2, y2)))

# Graficar las rectas
colores = plt.rcParams['axes.prop_cycle'].by_key()['color']
for i, (pendiente, termino_independiente, puntos_recta) in enumerate(zip(pendientes, terminos_independientes, puntos)):
    x = np.linspace(*rango_x, 100)
    y = pendiente * x + termino_independiente

    color = colores[i % len(colores)]
    plt.plot(x, y, color=color, label=f'Recta {i + 1}: Pendiente={pendiente:.2f}, Valor de la b={termino_independiente:.2f}')
    plt.scatter(*zip(*puntos_recta), color='red')

# Configuración del gráfico
plt.xlim(rango_x)
plt.ylim(rango_y)
plt.xticks(range(rango_x[0], rango_x[1] + 1))
plt.yticks(range(rango_y[0], rango_y[1] + 1))
plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Graficando rectas')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=1)

# Mostrar el gráfico
plt.show()
```
