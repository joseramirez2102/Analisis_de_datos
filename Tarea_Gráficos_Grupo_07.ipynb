## **FUNCIONES RETOS GRUPALES**

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Spotify_world_chart.csv')
data=pd.DataFrame(df)
    
#Ejercicio 1: columna vrs fila
col_name = input("Ingrese el nombre de la cancion: ")
col_name2 = input("Ingrese el nombre de columna: ")
    
value = data.loc[data['Título'] == col_name, col_name2].iloc[0]
print(f"Valor de {col_name2} para {col_name}: {value}")
print()

#Ejercicio 2: Ordenar la tabla por la columna "Semanas"
sorted_data = data.sort_values(by='Semanas', ascending=False)
print("Tabla ordenada por mayor cantidad de semanas:")
print(sorted_data)
print()

#Ejercicio 3: Mostrar fila por canción y artista ingresado por el usuario
col_name3 = input("Ingrese el nombre de la cancion: ")
col_name4 = input("Ingrese el nombre del artista: ")
    
value = data.loc[(data['Título']==col_name3)&(data['Artista']==col_name4)]
print(value)
print()

#Ejercicio 4: Obtener el valor total de una columna
column = input("Ingrese el nombre de la columna que quiere sumar: ")
suma = data[column].sum()
print(f"La suma de la columna {column} es: {suma}")

#Ejercicio 5: Una consulta con al menos 3 if
consulta = input("Ingrese su consulta (artista/titulo/posicion): ")

#Filtrar por artista
if "artista" in consulta.lower():
    nombre_artista = input("Ingrese el nombre del artista: ")
    tb = df[df['Artista'] == nombre_artista]

#Filtrar por titulo
if "titulo" in consulta.lower():
    nombre = input("Ingrese el nombre de la cancion: ")
    tb = df[df['Título'] == nombre]

#Ordenar por posicion ascendente o descendente
if "posicion" in consulta.lower():
    orden = input("Ingrese 'ascendente' o 'descendente': ")
    if orden == "ascendente":
        tb = df.sort_values('Posición', ascending=True)
    elif orden == "descendente":
        tb = df.sort_values('Posición', ascending=False)

print(tb)
```

## *Primer gráfico*

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Spotify_world_chart.csv')
data=pd.DataFrame(df)
#Este gráfico muestra las 25 canciones con mayor cantidad de semanas desde que fueron publicadas

#Seleccionamos las columnas a utilizar
graphic = df[['Título', 'Semanas']]

#Ordenamos los datos por la columna "Semanas"
graphic = graphic.sort_values(by='Semanas', ascending=False)

#Seleccionamos las primeras 25 canciones
graphic = graphic.head(25)

#Mostramos el gráfico
plt.barh(graphic['Título'], graphic['Semanas'], color='green')
plt.xticks(rotation=90)
plt.xlabel('Semanas', fontsize = 14)
plt.ylabel('Canciones', fontsize = 14)
plt.yticks(fontsize=8)
plt.show()
```

# *Segundo gráfico*

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Spotify_world_chart.csv')
data=pd.DataFrame(df)
# Agrupa el df por el artista y suma las veces que han estado en el top 10 y los ordena de mayor a menor
top10_por_artista = df.groupby('Artista')['Top 10'].sum().sort_values(ascending=False).head(25)

# crear un gráfico de barras horizontales para mostrar los artistas con más veces en el top 10
fig, ax = plt.subplots(figsize=(8, 10))

#Crea el tipo de gráfico y setea los colores
top10_por_artista.plot(kind='barh', ax=ax, color='royalblue')
ax.set_title('Los 10 artistas con más veces en el top 10')
ax.set_ylabel('Artista')
ax.set_xlabel('Número de veces en el top 10')
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Spotify_world_chart.csv')
data=pd.DataFrame(df)

# Aquí se ordena el DataFrame revisando las veces que una canción aparace en el Top 10
df_sorted = df.sort_values(by='Top 10', ascending=False)

# Se seleccionan las primeras 10 canciones
top_10_songs = df_sorted.head(10)

# Se crea un gráfico especificando que sea de tipo pastel
plt.pie(top_10_songs['Top 10'], labels=top_10_songs['Título'], autopct='%1.1f%%')
plt.title('Top 10 canciones según su número de apariciones en el Top 10 Global')
plt.axis('equal') 

# Mostrar el gráfico
plt.show()

#Tip: sacar un mapa de las localidades
```
